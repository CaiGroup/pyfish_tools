```python
lists = [np.random.randint(0, 800000, size=4) for _ in range(40000)]
sets = {frozenset(l): l for l in lists}
```

```python
values, indices, counts = np.unique(np.concatenate(lists), return_counts=True, return_index=True)
```

```python
bads = set(np.compress(counts>1, values))
```

```python
bads
```

```python
good_sets = set()
seen_bads = set()
for s in sets:
    if s & seen_bads:
        continue
    
    sb = s & bads
    if sb:
        seen_bads |= sb
    
    good_sets.add(s)
```
