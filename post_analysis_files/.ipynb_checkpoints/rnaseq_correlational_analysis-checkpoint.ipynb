```python
#convert data to pseudobulk rnaseq data
ch1_bulk = pd.DataFrame(ch1_mtx.mean(axis=1)).reset_index()
ch1_bulk.columns = ["Genes", "Counts"]
```

```python
comb_ch1 = pd.merge(rnaseq,ch1_bulk)
comb_ch1
```

```python
sort_ch1 = comb_ch1.sort_values(["FPKM"])
sort_ch1
```

```python
linreg = linregress(x = sort_ch1["FPKM"], y = sort_ch1["Counts"])
pearsonr = linreg[2]
pearsonr = round(pearsonr,3)
pearsonr
```

```python
plt.scatter(np.log10(sort_ch1["Counts"]),np.log10(sort_ch1["FPKM"]), s = 5, alpha=0.5)
plt.ylabel("Bulk RNAseq Log10(FPKM)", fontsize=12)
plt.xlabel("Pseudobulk Log10(Counts)", fontsize=12)
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12, rotation=0)
plt.annotate(f"Pearson's r= {pearsonr}", (-0.5,1.5), fontsize=12)
sns.despine()
```

```python
yodai_filt = yodai[["geneID","on"]]
yodai_filt.columns = ["Genes", "On"]
yt_lex = pd.merge(yodai_filt,ch1_bulk)
```

```python
yt_lex
```

```python
linreg = linregress(x = yt_lex["Counts"], y = yt_lex["On"])
pearsonr = linreg[2]
pearsonr = round(pearsonr,2)
pearsonr
```

```python
plt.scatter(np.log10(yt_lex["Counts"]),np.log10(yt_lex["On"]), s = 3, alpha=0.5)
plt.ylabel("Yodai's Counts", fontsize=12)
plt.xlabel("Lex's Counts", fontsize=12)
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12, rotation=0)
plt.annotate(f"Pearson's r= {pearsonr}", (-0.5,1.5), fontsize=12)
sns.despine()
```
