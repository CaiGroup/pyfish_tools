```python
import pandas as pd
from barcode_key_converter import *
```

```python
codebook = pd.read_csv("../../10k_barcodes_647.csv", index_col = 0)
```

```python
codebook
```

```python
#generate possible codes
from itertools import product
ps = np.arange(1,21,1)
perm_ps = list(product(ps,repeat=3))
```

```python
#does the lenght match
len(perm_ps) == 20**3
```

```python
#make list of codes that has been used
codes = codebook[["Hyb1","Hyb2","Hyb3"]].values
code_set = [tuple(i) for i in codes]
```

```python
#find difference
diff = list(set(perm_ps) - set(code_set))
len(diff)
```

```python
#get fake code with error correction
errorfakecode = []
for code in diff:
    error = (code[0]+code[1]-code[2]) % 20
    if error == 0:
        error = 20
    errorfakecode.append(error)
```

```python
#make fakebook
fakecodes = pd.DataFrame(diff)
fakecodes.columns = ["Hyb1","Hyb2","Hyb3"]
fakecodes["Hyb4"] = errorfakecode
```

```python
#make fake code names
fakegenes = []
for i in range(len(fakecodes)):
    fakegenes.append(f"fake{i}")
```

```python
#add fake names
fakecodes.index = fakegenes
```

```python
#add to codebook
codebook = pd.concat([codebook,fakecodes])
```

```python
codebook
```

```python
new_codebook = barcode_key_converter_within(codebook, num_hybs = 80, num_barcodes = 4, ch=1)
```

```python
new_codebook.iloc[0].values
```

```python
#write out new codebook
new_codebook.to_csv("codebook_string.csv")
```
